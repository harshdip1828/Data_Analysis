1.Reward the top performer (student) of each school based on cumulative marks scored in last three years for all the subjects

```python

files = {
    2019: '/content/Bangalore Schools 2019.xlsx',
    2020: '/content/Bangalore Schools 2020.xlsx',
    2021: '/content/Bangalore Schools 2021.xlsx',
}

data = []
for year, file_path in files.items():
    df = pd.read_excel(file_path)
    df.columns = df.columns.str.strip().str.lower()
    df['year'] = year
    data.append(df)

combined_data = pd.concat(data, ignore_index=True)

combined_data['total_marks'] = combined_data.iloc[:, 2:].sum(axis=1)

cumulative_marks = (
    combined_data.groupby(['student roll', 'student name'], as_index=False)
    .agg({'total_marks': 'sum'})
)

top_performer = cumulative_marks.loc[cumulative_marks['total_marks'].idxmax()]

print("Top Performer Over the Last Three Years:")
print(top_performer)
```

2.Rank each student within their own school based on their total marks scored in the year 2020 and compare the marks of Rank 10 for each school by arranging them in descending order

```python
s2020.columns = s2020.columns.str.strip().str.lower()

s2020['total_marks'] = s2020.iloc[:, 2:].sum(axis=1)

s2020['rank'] = s2020['total_marks'].rank(method='min', ascending=False)

rank_10_student = s2020[s2020['rank'] == 10]

ranked_students_sorted = s2020.sort_values(by='total_marks', ascending=False)

print("10th Ranked Student(s):")
print(rank_10_student[['student name', 'total_marks']])
print("\nTop Students Sorted by Total Marks:")
print(ranked_students_sorted[['student name', 'total_marks']].head(15))
```

3.Find out students with the highest improvement for each subject from 2019-21 combining all the schools together

```python
data = {}
for year, file_path in files.items():
    df = pd.read_excel(file_path)
    df.columns = df.columns.str.strip().str.lower()
    df['year'] = year
    data[year] = df

data_2019 = data[2019].set_index('student roll')
data_2021 = data[2021].set_index('student roll')

common_columns = list(set(data_2019.columns) & set(data_2021.columns))

numeric_columns = [col for col in common_columns if data_2019[col].dtype != object and data_2021[col].dtype != object]


improvement = data_2021[numeric_columns].subtract(data_2019[numeric_columns], fill_value=0)

improvement['student name'] = data_2019['student name']

top_improvers = {}
for subject in numeric_columns:
    if subject not in ['student name', 'year']:
        top_student = improvement[subject].idxmax()
        top_improvers[subject] = {
            'student roll': top_student,
            'student name': improvement.loc[top_student, 'student name'],
            'improvement': improvement.loc[top_student, subject],
        }

print("Students with the Highest Improvement for Each Subject (2019-2021):")
for subject, details in top_improvers.items():
    print(f"{subject.capitalize()}: {details}")
```

4.Identify best school for Arts, Science and Commerce streams based on marks scored by students in respective subjects for those streams in last three years

```python
file_paths = {
    "2021": "/content/Bangalore Schools 2021.xlsx",
    "2020": "/content/Bangalore Schools 2020.xlsx",
    "2019": "/content/Bangalore Schools 2019.xlsx",
}

def compute_stream_averages(data):
    arts_subjects = ['History', 'Geography', 'Civics']
    science_subjects = ['Mathematics', 'Physics', 'Chemistry', 'Biology']
    commerce_subjects = ['Computer Science']

    arts_avg = data[arts_subjects].mean(axis=1).mean() if all(sub in data for sub in arts_subjects) else None
    science_avg = data[science_subjects].mean(axis=1).mean() if all(sub in data for sub in science_subjects) else None
    commerce_avg = data[commerce_subjects].mean(axis=1).mean() if all(sub in data for sub in commerce_subjects) else None

    return {"Arts": arts_avg, "Science": science_avg, "Commerce": commerce_avg}

# Aggregate results for all years
results = {}
for year, path in file_paths.items():
    year_data = pd.ExcelFile(path)
    year_results = {}
    for sheet in year_data.sheet_names:
        school_data = pd.read_excel(year_data, sheet_name=sheet)
        year_results[sheet] = compute_stream_averages(school_data)
    results[year] = year_results

# Print the results for each year
for year, schools in results.items():
    print(f"Year: {year}")
    for school, averages in schools.items():
        print(f"  {school}:")
        for stream, avg in averages.items():
            print(f"    {stream}: {avg:.2f}" if avg is not None else f"    {stream}: No data")
    print("\n")
```

Then calculate for each school how many students were in each category based on the avg. marks obtained each year

```python


# Function to compute average marks per stream for a given dataset
def compute_stream_averages(data):
    arts_subjects = ['History', 'Geography', 'Civics']
    science_subjects = ['Mathematics', 'Physics', 'Chemistry', 'Biology']
    commerce_subjects = ['Computer Science']

    arts_avg = data[arts_subjects].mean(axis=1).mean() if all(sub in data for sub in arts_subjects) else None
    science_avg = data[science_subjects].mean(axis=1).mean() if all(sub in data for sub in science_subjects) else None
    commerce_avg = data[commerce_subjects].mean(axis=1).mean() if all(sub in data for sub in commerce_subjects) else None

    return {"Arts": arts_avg, "Science": science_avg, "Commerce": commerce_avg}

# Function to categorize students based on their average marks
def categorize_students(data):
    categories = {
        "Very Poor": (0, 20),
        "Poor": (21, 40),
        "Average": (41, 60),
        "Good": (61, 80),
        "Very Good": (81, 100)
    }

    avg_marks = data.mean(axis=1)
    category_counts = {category: 0 for category in categories}

    for mark in avg_marks:
        for category, (low, high) in categories.items():
            if low <= mark <= high:
                category_counts[category] += 1
                break

    return category_counts

# Aggregate results for all years
results = {}
category_results = {}

for year, path in file_paths.items():
    year_data = pd.ExcelFile(path)
    year_results = {}
    year_category_results = {}

    for sheet in year_data.sheet_names:
        school_data = pd.read_excel(year_data, sheet_name=sheet)
        year_results[sheet] = compute_stream_averages(school_data)
        year_category_results[sheet] = categorize_students(school_data.iloc[:, 2:])  # Exclude non-subject columns

    results[year] = year_results
    category_results[year] = year_category_results

# Print the results for each year
for year, schools in results.items():
    print(f"Year: {year}")
    for school, averages in schools.items():
        print(f"  {school}:")
        for stream, avg in averages.items():
            print(f"    {stream}: {avg:.2f}" if avg is not None else f"    {stream}: No data")
    print("\n")

# Print the category distribution for each year
for year, schools in category_results.items():
    print(f"Year: {year} - Category Distribution")
    for school, categories in schools.items():
        print(f"  {school}:")
        for category, count in categories.items():
            print(f"    {category}: {count}")
    print("\n")
```

7.Which is the fastest-growing School in Bangalore (Overall and Streamwise)?

```python


# Function to compute average marks per stream for a given dataset
def compute_stream_averages(data):
    arts_subjects = ['History', 'Geography', 'Civics']
    science_subjects = ['Mathematics', 'Physics', 'Chemistry', 'Biology']
    commerce_subjects = ['Computer Science']

    arts_avg = data[arts_subjects].mean(axis=1).mean() if all(sub in data for sub in arts_subjects) else None
    science_avg = data[science_subjects].mean(axis=1).mean() if all(sub in data for sub in science_subjects) else None
    commerce_avg = data[commerce_subjects].mean(axis=1).mean() if all(sub in data for sub in commerce_subjects) else None

    return {"Arts": arts_avg, "Science": science_avg, "Commerce": commerce_avg}

# Function to categorize students based on their average marks
def categorize_students(data):
    categories = {
        "Very Poor": (0, 20),
        "Poor": (21, 40),
        "Average": (41, 60),
        "Good": (61, 80),
        "Very Good": (81, 100)
    }

    avg_marks = data.mean(axis=1)
    category_counts = {category: 0 for category in categories}

    for mark in avg_marks:
        for category, (low, high) in categories.items():
            if low <= mark <= high:
                category_counts[category] += 1
                break

    return category_counts

# Aggregate results for all years
results = {}
category_results = {}

for year, path in file_paths.items():
    year_data = pd.ExcelFile(path)
    year_results = {}
    year_category_results = {}

    for sheet in year_data.sheet_names:
        school_data = pd.read_excel(year_data, sheet_name=sheet)
        year_results[sheet] = compute_stream_averages(school_data)
        year_category_results[sheet] = categorize_students(school_data.iloc[:, 2:])  # Exclude non-subject columns

    results[year] = year_results
    category_results[year] = year_category_results

# Print the results for each year
for year, schools in results.items():
    print(f"Year: {year}")
    for school, averages in schools.items():
        print(f"  {school}:")
        for stream, avg in averages.items():
            print(f"    {stream}: {avg:.2f}" if avg is not None else f"    {stream}: No data")
    print("\n")

# Print the category distribution for each year
for year, schools in category_results.items():
    print(f"Year: {year} - Category Distribution")
    for school, categories in schools.items():
        print(f"  {school}:")
        for category, count in categories.items():
            print(f"    {category}: {count}")
    print("\n")

# Determine the best school for each year based on Good and Very Good categories
best_schools = {}

for year, schools in category_results.items():
    best_school = None
    max_good_very_good = 0

    for school, categories in schools.items():
        good_very_good_total = categories.get("Good", 0) + categories.get("Very Good", 0)

        if good_very_good_total > max_good_very_good:
            best_school = school
            max_good_very_good = good_very_good_total

    best_schools[year] = (best_school, max_good_very_good)

# Print the best school for each year
print("Best Schools by Year based on Good and Very Good categories:")
for year, (school, count) in best_schools.items():
    print(f"Year {year}: {school} with {count} students in Good and Very Good categories")

# Determine the fastest-growing school (overall and streamwise)
def compute_growth(school_results):
    growth = {}

    for school in school_results["2019"].keys():
        growth[school] = {}

        for stream in ["Arts", "Science", "Commerce"]:
            try:
                start = school_results["2019"].get(school, {}).get(stream, 0)
                end = school_results["2021"].get(school, {}).get(stream, 0)

                if start is not None and end is not None:
                    growth[school][stream] = end - start
                else:
                    growth[school][stream] = None
            except KeyError:
                growth[school][stream] = None

        # Compute overall growth
        overall_start = sum(filter(None, [
            school_results["2019"].get(school, {}).get(stream, 0)
            for stream in ["Arts", "Science", "Commerce"]
        ]))

        overall_end = sum(filter(None, [
            school_results["2021"].get(school, {}).get(stream, 0)
            for stream in ["Arts", "Science", "Commerce"]
        ]))

        growth[school]["Overall"] = overall_end - overall_start

    return growth

school_growth = compute_growth(results)

# Identify the fastest-growing school (overall and streamwise)
fastest_growing = {}

for stream in ["Arts", "Science", "Commerce", "Overall"]:
    best_school = None
    max_growth = float("-inf")

    for school, growth in school_growth.items():
        if growth.get(stream, float("-inf")) > max_growth:
            best_school = school
            max_growth = growth[stream]

    fastest_growing[stream] = (best_school, max_growth)

# Print the fastest-growing schools
print("Fastest-Growing Schools (Overall and Streamwise):")
for stream, (school, growth) in fastest_growing.items():
    print(f"{stream}: {school} with a growth of {growth}")
```
